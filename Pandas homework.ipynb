```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(12,6))
imdb['Rating'].plot.hist(ax=axes[0], bins=80, alpha=0.5, color='red')
axes[0].set_title('Rating')

imdb['Meta Score'].plot.hist(ax=axes[1], bins=80, alpha=.5, color='blue')
axes[1].set_title('Meta Score')

imdb['Votes'].plot.hist(ax=axes[2], bins=80, alpha=0.5, color='green')
axes[2].set_title('Votes')

axes[1].set_ylabel('')
axes[2].set_ylabel('')
```
